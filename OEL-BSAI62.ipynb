```python
def string_pattern(s, p):

    words = s.split()


    if len(words) != len(p):
        return False

    map_s_to_p = {}
    map_p_to_s = {}

    for char_s, char_p in zip(words, p):

        if char_s in map_s_to_p and map_s_to_p[char_s] != char_p:
            return False

        if char_p in map_p_to_s and map_p_to_s[char_p] != char_s:
            return False

        map_s_to_p[char_s] = char_p
        map_p_to_s[char_p] = char_s

    return True

def print_output():
    print(string_pattern("dog cat cat dog", "abba")) # True
    print(string_pattern("dog cat cat fish", "abba")) # False
    print(string_pattern("dog cat cat dog", "aaaa")) # False

print_output()
```
